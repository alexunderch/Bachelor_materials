```python
import numpy as np
from scipy import linalg
import matplotlib
from matplotlib import pylab as plt
```

```python
%matplotlib inline

x = np.arange(1, 16,0.1)
f = np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp (-x / 2)
plt.plot(x, f)
```

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp (-x / 2)
    
```

```python
A = np.array([[1, 1], [1, 15]])
b = np.array ([f(1), f(15)])
x = linalg.solve(A, b)
print (x)
```

```python
plt.plot(x, f(x))
```

```python
A = np.array([[1, 1, 1], [1, 8, 8 ** 2], [1, 15, 15 ** 2]])
b = np.array ([f(1), f(8), f(15)])
x = linalg.solve(A, b)
print (x)
```

```python
plt.plot(x, f(x))
```

```python
A = np.array([[1, 1, 1, 1], [1, 4, 16, 64], [1, 10, 100, 1000], [1, 15, 15 ** 2, 15 ** 3]])
b = np.array ([f(1), f(4), f(10), f(15)])
x = linalg.solve(A, b)
print (x)
```

```python

plt.plot(x, f(x))
print (np.round(x,2))
```
