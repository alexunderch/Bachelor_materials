```python
import scipy.stats as sts
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Экспоненциальное распределение ($\rho = \lambda e^{-\lambda x}$, $\lambda =1$) и его плотность

```python
x = np.linspace(-1, 10, 1000)
pdf = sts.expon.pdf(x)
cdf = sts.expon.cdf(x)

plt.plot(x, pdf, label ='Theoretical PDF')
plt.plot(x, cdf, label ='Theoretical CDF')

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc ='lower right')
```

# Построение гистограммы по выорке объёма 1000

```python
x = np.linspace(-0.1, 8, 1000)
pdf = sts.expon.pdf(x)

exp_rv = sts.expon(0)
sample = exp_rv.rvs(1000)

plt.hist(sample, normed=True)
plt.plot(x, pdf, color = 'k')

plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

## Сгенереруем 1000 выборок объёма n 

## Функция, генерирующая выборку из 1000 выборочных средних

```python
def generator(n):
    x = 1
    smpl = np.array([])
    while x <= 1000:
        exp_rv = sts.expon(0)
        sample = exp_rv.rvs(n)
        mean = sum(sample)/n
        smpl = np.append(smpl, mean)
        x+=1
    return smpl
```

## $E\xi$ = $\frac{1}{\lambda}$, $D\xi$ = $\frac{1}{\lambda^2}$ => $\sigma = \sqrt{\frac{1}{n\lambda}}$, где $\xi$ - случайная величина, распределённая экспоненциально с параметром $\lambda$ = 1

## n = 5, $E\xi$ = 1, $\sigma = \sqrt{0.2}=0.45$

```python
norm = sts.norm(1, np.sqrt(1./5))
x = np.linspace(0, 3, 1000)
pdf = norm.pdf(x)

plt.plot(x, pdf, color = 'k')
plt.hist(generator(5),bins =10, range=(0,3), normed=True)
plt.ylabel('number of samples')
plt.xlabel('$x$')
```
