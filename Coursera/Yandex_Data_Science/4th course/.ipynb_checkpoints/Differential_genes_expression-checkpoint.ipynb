```python
cancer = data[data["Diagnosis"] == 'cancer']
early_neoplasia = data[data["Diagnosis"] == 'early neoplasia']
normal = data[data["Diagnosis"] == 'normal']
```

```python
normal_vs_early_neoplasia = {}
cancer_vs_early_neoplasia = {}
genes = data.columns[2:]
for gene in genes:
    normal_vs_early_neoplasia[gene] = ttest_ind(normal[gene], early_neoplasia[gene], equal_var = False).pvalue
    cancer_vs_early_neoplasia[gene] = ttest_ind(cancer[gene], early_neoplasia[gene], equal_var = False).pvalue 
```

```python
normal_vs_early_neoplasia_df = pd.DataFrame.from_dict(normal_vs_early_neoplasia, 
                                                      orient = 'index', columns = ['p-Value_norm_vs_neo'])
cancer_vs_early_neoplasia_df = pd.DataFrame.from_dict(cancer_vs_early_neoplasia, 
                                                      orient = 'index', columns = ['p-Value_cancer_vs_neo'])
```

```python
print("Statistically valuable in normal vs early neoplasia: ",
      normal_vs_early_neoplasia_df[normal_vs_early_neoplasia_df["p-Value_norm_vs_neo"] < 0.05].shape[0])
print("Statistically valuable in normal vs early neoplasia: ",
      cancer_vs_early_neoplasia_df[cancer_vs_early_neoplasia_df["p-Value_cancer_vs_neo"] < 0.05].shape[0])

answers.append(normal_vs_early_neoplasia_df[normal_vs_early_neoplasia_df["p-Value_norm_vs_neo"] < 0.05].shape[0])
answers.append(cancer_vs_early_neoplasia_df[cancer_vs_early_neoplasia_df["p-Value_cancer_vs_neo"] < 0.05].shape[0])
```
