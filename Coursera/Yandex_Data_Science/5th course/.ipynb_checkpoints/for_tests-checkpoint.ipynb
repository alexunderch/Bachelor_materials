```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
```

```python
milk = pd.read_csv('monthly-milk-production.csv',';',
                   index_col = ['month'], parse_dates = ['month'], dayfirst = True)
```

```python
milk.head()
```

```python
plt.plot(milk.index, milk.milk)
```

```python
import statsmodels.api as sm
sm.tsa.stattools.adfuller(milk.milk)[1]
```

```python
milk['milk_daily'] = milk.milk / milk.index.days_in_month
```

```python
plt.plot(milk.index, milk.milk_daily)
```

```python
round(sum(milk.milk_daily), 2)
```

```python
milk['daily_diff1'] = milk.milk_daily - milk.milk_daily.shift(1)
milk.dropna(inplace = True)
sm.tsa.stattools.adfuller(milk.daily_diff1)[1]
plt.plot(milk.daily_diff1)
```

```python
milk['daily_diff12'] = milk.milk_daily - milk.milk_daily.shift(12)
milk.dropna(inplace = True)
sm.tsa.stattools.adfuller(milk.daily_diff12)[1]
plt.plot(milk.daily_diff12)
```

```python
milk['daily_diff2'] = milk.daily_diff12 - milk.daily_diff12.shift(1)
milk.dropna(inplace = True)
sm.tsa.stattools.adfuller(milk.daily_diff2)[1]
plt.plot(milk.daily_diff2)
```

```python
import statsmodels.api as sm
sm.graphics.tsa.plot_acf(milk.daily_diff2.values.squeeze(), lags=50)
```

```python
import statsmodels.api as sm
sm.graphics.tsa.plot_pacf(milk.daily_diff2.values.squeeze(), lags = 50)
```
