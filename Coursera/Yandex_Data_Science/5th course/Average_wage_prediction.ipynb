# Прогнозирование уровня средней заработной платы в России

```python
# импортируем необходимые библиотеки
%pylab inline
import pandas as pd
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
import warnings
from itertools import product

def invboxcox(y,lmbda):
    if lmbda == 0:
        return(np.exp(y))
    else:
        return(np.exp(np.log(lmbda*y+1)/lmbda))
```

Загрузим данные и посмотрим на них


```python
wage = pd.read_csv('WAG_C_M.csv',';', index_col = ['month'], parse_dates = ['month'], dayfirst=True)
wage.columns = ['amount']
wage.amount = wage.amount * 1000
plt.figure(figsize(15, 7))
wage.amount.plot()
plt.ylabel('Wage')
pylab.show()
```

При визуальном анализе видно, что дисперсия ряда не стационарна, нужна её стабилизация с помощью *преобразования Бокса-Кокса* 

```python
wage['amount_box'], lmbda = stats.boxcox(wage.amount)
plt.figure(figsize(15, 7))
wage.amount_box.plot()
plt.ylabel(u'Transformed wage amount')
```

Так гораздо лучше. Проверим ряд на стационарность и сделаем STL-декомпозицию.

```python
plt.figure(figsize(15, 10))
sm.tsa.seasonal_decompose(wage.amount_box).plot()
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(wage.amount_box)[1])
```

Ряд нестационарный с ярко выраженным возрастающим трендом и сезонностью. Проведём сезонное дифференцирование и посмотрим, что изменится

```python
wage['amount_box_diff'] = wage.amount_box - wage.amount_box.shift(12)
plt.figure(figsize(15, 10))
sm.tsa.seasonal_decompose(wage.amount_box_diff[12:]).plot()
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(wage.amount_box_diff[12:])[1])
```

Критерий Дики-Фуллера нестационарность уверенно отвергает, хотя тут что-то не так с трендом, выглядит не сильно стационарно. Продифференцируем ещё раз


```python
wage['amount_box_diff2'] = wage.amount_box_diff - wage.amount_box_diff.shift(1)
plt.figure(figsize(15, 10))
sm.tsa.seasonal_decompose(wage.amount_box_diff2[13:]).plot()   
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(wage.amount_box_diff2[13:])[1])
```

Несмотря на пару горбиков, ряд выглядит визуально лучше $-$ большая его часть ведёт себя похоже на константу.

С другой стороны, в сезонно продифференцированном ряду гипотеза нестационарности также уверенно отвергается. Возьмём с собой в следующий этап две модели.
* D = 1, d = 1
* D = 1, d = 0

## Подбор модели


Построим полную и частичную автокоррелограммы нашего ряда

```python
plt.figure(figsize(15, 8))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(wage.amount_box_diff2[13:].values.squeeze(), lags = 48, ax=ax)
pylab.show()
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(wage.amount_box_diff2[13:].values.squeeze(), lags = 48, ax=ax)
pylab.show()
```

Из графиков возьмём так:
* Q = 0
* q = 1
* P = 4
* p = 5

```python
ps = range(0, 6)
d = 1
qs = range(0, 2)
Ps = range(0, 5)
D = 1
Qs = range(0, 2)

parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
len(parameters_list)
```

Обучение моделей и выбор лучшей:

```python
%%time
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')

for param in parameters_list:
    #try except нужен, потому что на некоторых наборах параметров модель не обучается
    try:
        model = sm.tsa.statespace.SARIMAX(wage.amount_box, order = (param[0], d, param[1]), 
                                        seasonal_order = (param[2], D, param[3], 12)).fit(disp=-1)
    #выводим параметры, на которых модель не обучается и переходим к следующему набору
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    #сохраняем лучшую модель, aic, параметры
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
    
warnings.filterwarnings('default')
```

```python
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending = True).head())
```

```python
print(best_model.summary())
```

Скорее всего, модель не автокоррелирована.
Проведём анализ остатков, проверим:

```python
plt.figure(figsize(15,8))
plt.subplot(211)
best_model.resid[13:].plot()
plt.ylabel(u'Residuals')

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(best_model.resid[13:].values.squeeze(), lags=48, ax=ax)

print("Критерий Стьюдента: p=%f" % stats.ttest_1samp(best_model.resid[13:], 0)[1])
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(best_model.resid[13:])[1])
```

Остатки несмещены (подтверждается критерием Стьюдента) стационарны (подтверждается критерием Дики-Фуллера и визуально), неавтокоррелированы (подтверждается критерием Льюнга-Бокса и коррелограммой). Посмотрим, насколько хорошо модель описывает данные:

```python
wage['model'] = invboxcox(best_model.fittedvalues, lmbda)
plt.figure(figsize(15,7))
wage.amount.plot()
wage.model[13:].plot(color='r')
plt.ylabel('Average Wage Amount')
pylab.show()
```
